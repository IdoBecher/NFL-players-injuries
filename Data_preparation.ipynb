**Cleaning**

```python
InjuryRecord = InjuryRecord.sort_values('PlayerKey').dropna()
```

```python
InjuryRecord.reset_index(inplace=True)
```

```python
InjuryRecord.drop(columns='index', inplace=True)
InjuryRecord.head()
```

```python
# Getting the data of the injured players from PlayerTrackData
l = len(PlayerTrackData)
ind = []
mid = 0
for i in InjuryRecord.index:
  keyI = InjuryRecord['PlayKey'][i][:5]
  #print(keyI)
  high = int(l-210000)
  low = mid + 1
  while(high>=low):
    mid = int((high + low)/2)
    keyD = PlayerTrackData['PlayKey'][mid][:5]
    if keyD == keyI:
      ind.append(mid)
      #print(mid)
      mid2 = mid
      mid3 = mid
      while PlayerTrackData['PlayKey'][mid2+1][:5]==keyI:
        mid2+=1
        ind.append(mid2)
        #print(mid2)
      while PlayerTrackData['PlayKey'][mid3-1][:5]==keyI:
        mid3-=1
        ind.append(mid3)
        #print(mid3)

      break
    elif int(keyD) > int(keyI):
      high = mid - 1
    else:
      low = mid +1
```

```python
ind.sort()
```

```python
D = pd.DataFrame(data={'Key': PlayerTrackData['PlayKey'][ind], 'time': PlayerTrackData['time'][ind], 'event': PlayerTrackData['event'][ind], \
                       'x': PlayerTrackData['x'][ind], 'y': PlayerTrackData['y'][ind], 'o': PlayerTrackData['o'][ind], 's': PlayerTrackData['s'][ind], \
                       'dir': PlayerTrackData['dir'][ind], 'dis': PlayerTrackData['dis'][ind]})
```

```python
D.head()
```

```python
D2 = D.reset_index().drop(columns=['index'])
D2.head()
```

```python
print(f'The length of D2 is {len(D2)}')
```

**Data reduction**

```python
# option 1
# reduction per attack
deltaT = []
deltaX = []
deltaY = []
TopSpeed = []
avgSpeed = []
maxAngle = []
keyP = []
t=0
d=D2['dis'][0]
sp = 0
angle = 0
distance = []
for i in D2.index+1:
  if D2['Key'][i-1][-2:]==D2['Key'][i][-2:]:
    x = min(abs(360-D2['dir'][i-1]+D2['dir'][i]), abs(360-D2['dir'][i]+D2['dir'][i-1]), abs(D2['dir'][i-1]-D2['dir'][i]))
    if angle < x:
      angle = x
    if sp < D2['s'][i]:
      sp = D2['s'][i]
    t+=D2['time'][i]-D2['time'][i-1]
    d+=abs(D2['dis'][i-1]-D2['dis'][i])
  else:
    #print(D2['Key'][i-1])
    keyP.append(D2['Key'][i-1])
    TopSpeed.append(sp)
    sp = 0
    maxAngle.append(angle)
    angle = 0
    deltaT.append(t)
    distance.append(d)
    t=0
    d=D2['dis'][i]
  if i==D2.index[-1]:
    break
```

```python
# option 2
# reduction per event  
deltaT = []
deltaX = []
deltaY = []
TopSpeed = []
avgSpeed = []
maxAngle = []
keyP = []
t=0
d=D2['dis'][0]
sp = 0
angle = 0
distance = []
for i in D2.index+1:
  x = min(abs(360-D2['dir'][i+1]+D2['dir'][i]), abs(360-D2['dir'][i]+D2['dir'][i+1]), abs(D2['dir'][i+1]-D2['dir'][i]))
  if pd.isna(D2['event'][i]):
    if angle < x:
      angle = x
    if sp < D2['s'][i]:
      sp = D2['s'][i]
    t+=D2['time'][i+1]-D2['time'][i]
    d+=abs(D2['dis'][i+1]-D2['dis'][i])
  else:
    keyP.append(D2['Key'][i-1])
    TopSpeed.append(sp)
    sp = 0
    maxAngle.append(angle)
    angle = 0
    deltaT.append(t)
    distance.append(d)
    t=0
    d=D2['dis'][i]
  if i==D2.index[-2]:
    break
```

```python
keyP.append(D2['Key'][i-1])
TopSpeed.append(sp)
maxAngle.append(angle)
deltaT.append(t)
distance.append(d)
```

```python
I = pd.DataFrame(data={'Key': keyP, 'time': deltaT, 'dis': distance, 'TopSpeed': TopSpeed, 'maxAngle': maxAngle})
```

**Body part column**

```python
bp = []
#sur = []
s=0
for k in I2.index:
  flag = 0
  for j in InjuryRecord.index:
    if I2['Key'][k] == InjuryRecord['PlayKey'][j]:
      #if I2['Key'][k][:8] == InjuryRecord['PlayKey'][j][:8]:
        #sur.append(InjuryRecord['Surface'][j])
      flag = 1
      s+=1
      break
    
  if flag == 1:
    bp.append(InjuryRecord['BodyPart'][j])
    
  else:
    bp.append('health')
```

```python
# Getting the data of the injured players from PlayerTrackData
ind2 = []
mid = 0
for i in InjuryRecord.index:
  keyI = InjuryRecord['PlayKey'][i][:5]
  #print(keyI)
  high = len(PlayList)
  low = 0
  while(high>=low):
    mid = int((high + low)/2)
    keyD = PlayList['PlayKey'][mid][:5]
    if keyD == keyI:
      
      ind2.append(mid)
      #print(mid)
      mid2 = mid
      mid3 = mid
      while PlayList['PlayKey'][mid2+1][:5]==keyI:
        mid2+=1
        ind2.append(mid2)
        #print(mid2)
      while PlayList['PlayKey'][mid3-1][:5]==keyI:
        mid3-=1
        ind2.append(mid3)
        #print(mid3)
      
      #df['PlayKey'][mid] = '0'
      break
    elif int(keyD) > int(keyI):
      high = mid - 1
    else:
      low = mid +1
```

```python
PlayList2 = pd.DataFrame(data={'PlayKey': PlayList['PlayKey'][ind2], 'RosterPosition': PlayList['RosterPosition'][ind2], \
                               'FieldType': PlayList['FieldType'][ind2], 'PlayType': PlayList['PlayType'][ind2]})
```

```python
(PlayList2.reset_index(inplace=True))
PlayList2.drop(columns='index', inplace=True)
PlayList2.head()
```

```python
i, j = 0, 0
FieldType = []
RosterPosition = []
PlayType = []
while(j < len(PlayList2) and i < len(I2)):
  #keyI = I2['Key']
  #keypl = PlayList['PlayKey']
  while (i < len(I2) and I2['Key'][i] == PlayList2['PlayKey'][j]):
    FieldType.append(PlayList2['FieldType'][j])
    RosterPosition.append(PlayList2['RosterPosition'][j])
    PlayType.append(PlayList2['PlayType'][j])
    i+=1
  j+=1
```
